```python
import cv2
import matplotlib.pyplot as plt

img=cv2.imread("/content/WhatsApp Image 2025-12-09 at 10.38.27 AM.jpeg",0)
heimg=cv2.equalizeHist(img)
plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
plt.title("Original Image")
plt.imshow(img,cmap="gray")
plt.axis("Off")

plt.subplot(1,3,2)
plt.title("Histogram Equalization")
plt.imshow(heimg,cmap="gray")
plt.axis("Off")

plt.subplot(1,3,3)
plt.title("Histogram(Equlaized)")
plt.hist(heimg.ravel(),256)
plt.show()
```
